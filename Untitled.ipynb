```python
import pandas as pd
```

```python
contact_preference = pd.read_csv('data/contact_preference.csv', delimiter = ';')
customer_preferenced_contact_type = pd.read_csv('data/customer_preferenced_contact_type.csv', delimiter = ';')
customer = pd.read_csv('data/customer.csv', delimiter = ';')
customers_segmetns = pd.read_csv('data/customers_segmetns.csv', delimiter = ';')
location = pd.read_csv('data/location.csv', delimiter = ';')
product_feedback = pd.read_csv('data/product_feedback.csv', delimiter = ';')
product = pd.read_csv('data/product.csv', delimiter = ';')
purchase = pd.read_csv('data/purchase.csv', delimiter = ';')
segment = pd.read_csv('data/segment.csv', delimiter = ';')
service_feedback = pd.read_csv('data/service_feedback.csv', delimiter = ';')
```

## Task1

```python
orders_amount = purchase.groupby('user_id').count().rename(columns={'id':'count'})[['count']]

purhcase_with_products = purchase.merge(product, left_on = 'product_id', right_on='id')
purhcase_with_products['price'] = pd.to_numeric(purhcase_with_products['price'].str.replace('Free', '0'))
purhcase_with_products['total'] = purhcase_with_products['quantity'] * purhcase_with_products['price']
spent = purhcase_with_products[['user_id', 'total']].groupby('user_id').sum()


mean_user_product_rating = purchase.merge(product_feedback, left_on='id', right_on='purchase_id')[['user_id', 'rating', 'product_id']] \
    .groupby(['user_id', 'product_id']).mean()[['rating']].reset_index()
max_idx = mean_user_product_rating.groupby('user_id')['rating'].transform('max') == mean_user_product_rating['rating']
favourite_user_product = mean_user_product_rating[max_idx].groupby('user_id').first()[['product_id']]

mean_user_category_rating = purchase.merge(product_feedback, left_on='id', right_on='purchase_id')[['user_id', 'rating', 'product_id']] \
    .merge(product, left_on='product_id', right_on='id')[['user_id', 'category_name', 'rating']] \
    .groupby(['user_id', 'category_name']).mean()[['rating']].reset_index()
max_idx = mean_user_product_rating.groupby('user_id')['rating'].transform('max') == mean_user_product_rating['rating']
favourite_user_category = mean_user_category_rating[max_idx].groupby('user_id').first()[['category_name']]

feedback_count = pd.concat([
    product_feedback.merge(purchase, left_on = 'purchase_id', right_on = 'id')[['user_id', 'rating']],
    service_feedback[['customer_id', 'rating']].rename(columns = {'customer_id':'user_id'})
]).groupby('user_id').count().rename(columns = {'rating':'feedback_count'})

result = customer \
    .merge(orders_amount, left_on='id', right_on='user_id').rename(columns = {'count':'orders_amount'}) \
    .merge(spent, left_on='id', right_on='user_id') \
    .merge(favourite_user_product, left_on='id', right_on='user_id').rename(columns = {'product_id':'fav_product'}) \
    .merge(favourite_user_category, left_on='id', right_on='user_id').rename(columns = {'category_name':'fav_category'}) \
    .merge(feedback_count, left_on='id', right_on='user_id')

result
```
